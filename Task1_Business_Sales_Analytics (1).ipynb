```python
import pandas as pd

df = pd.read_csv('Sample_ Superstore.csv')

print("Dataset loaded successfully!")
print("Rows and Columns:", df.shape)
```

```python
import matplotlib.pyplot as plt

sales_by_category = df.groupby('Category')['Sales'].sum()

sales_by_category.plot(kind='bar')

plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')

plt.show()
```

```python
profit_by_region = df.groupby('Region')['Profit'].sum()

profit_by_region.plot(kind='bar')

plt.title('Profit by Region')
plt.xlabel('Region')
plt.ylabel('Total Profit')

plt.show()
```

```python
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()

print("Total Sales =", total_sales)
print("Total Profit =", total_profit)
```

```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
top_products.plot(kind='bar')

plt.title('Top 10 Products by Sales')
plt.xlabel('Product Name')
plt.ylabel('Sales')

plt.show()
```

```python
profit_category = df.groupby('Category')['Profit'].sum()

profit_category.plot(kind='bar')

plt.title('Profit by Category')
plt.xlabel('Category')
plt.ylabel('Profit')

plt.show()
```

```python
region_sales = df.groupby('Region')['Sales'].sum()

plt.figure(figsize=(6,4))
region_sales.plot(kind='bar')
plt.title('Sales by Region')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.show()
```

Business Recommendations

1. Technology products generate the highest profit and should receive greater marketing focus.

2. The West region is the most profitable region and can be used as a benchmark for other regions.

3. Furniture has lower profitability compared to its sales volume. Pricing and discount strategies should be reviewed.

4. The Canon imageCLASS 2200 Advanced Copier is the highest-selling product and should be prioritized for inventory management.

5. The company should focus on high-profit products and regions to maximize overall profitability.

Conclusion

The analysis revealed that Technology is the most profitable category, while the West region generates the highest sales and profit.
